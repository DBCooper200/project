```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("income.csv")
df
```

```python
plt.scatter(df['Age'],df['Income($)'],color="pink")
#we observe 3 clusters visually so we can easily say k=3 i.e. 3 centroids
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
#clusters have been formed and they have been labelled as 0,1,2
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color="red",label="df1")
plt.scatter(df2.Age,df2['Income($)'],color="pink",label="df2")
plt.scatter(df3.Age,df3['Income($)'],color="magenta",label="df3")
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

scaling of feature is improper
y has wide range but x is narrow thus we need minmax preprocessing 

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df.Age=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df["cluster"]=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color="red",label="df1")
plt.scatter(df2.Age,df2['Income($)'],color="pink",label="df2")
plt.scatter(df3.Age,df3['Income($)'],color="magenta",label="df3")
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color="red",label="df1")
plt.scatter(df2.Age,df2['Income($)'],color="pink",label="df2")
plt.scatter(df3.Age,df3['Income($)'],color="magenta",label="df3")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",label="centroid")
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_) #inertia gives sum of square error
```

```python
sse
```

```python
plt.xlabel("k")
plt.ylabel("sse")
plt.plot(k_rng,sse,color="pink")
#we can manually see the elbow point at k=3
```
