```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('homeprices.csv')
```

```python
df
```

```python
pd.get_dummies(df.town)
```

```python
dummies=pd.get_dummies(df.town)
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged #we can drop town also
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop(['price'],axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[3400,1,0]])
```

```python
#to check how accurate the predictions are we use score() it predicts prices for all values in x on its own and the compares with y
model.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
X=dfle[['town','area']].values
#values makes x a 2d array and not a dataframe
X
```

```python
y=dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer
ohe=ColumnTransformer([("town", OneHotEncoder(), [0])], remainder = 'passthrough')#0th column of x is a categorical variable
```

```python
X=np.array(ohe.fit_transform(X))
X
```

```python
#removing one dummy variable
x=X[:,1:]
x
```

```python
model1=LinearRegression()
model1.fit(x,y)
```

```python
model1.predict([[1,0,2800]])
```

```python
model1.score(x,y)
```
